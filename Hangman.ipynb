```python
# Hangman
# Tanner Atkin
# 10/25/2022

import random

# variables
wordBank = ["cougar", "school", "awesome", "halloween", "favorite",
              "magenta", "doughnut", "apples", "bean", "celery"]
gameContinue = True

# start game
while gameContinue:

  # reset variables
  guessWord = random.choice(wordBank).upper()
  correctLetters = ["_"] * len(guessWord)
  incorrectLetters = []
  numCorrectGuesses = 0
  numIncorrectGuesses = 0
  guessContinue = True

  # welcome user
  guessContinue = True

  # welcome user
  print("\nWelcome to Hangman!\n")
  print(f"There are {len(guessWord)} letters in your word.")
  print("_ " * len(guessWord) + "\n")

  # user begins guessing
  while guessContinue:
    guessLetter = input("Guess a letter:\t").upper()
    # check length
    if len(guessLetter) <= 1:
      # check if alphabetical
      if guessLetter.isalpha():
        # check if already guessed
        if not (guessLetter in correctLetters or guessLetter in incorrectLetters):
          # check if correct or incorrect
          if guessLetter in guessWord:
            print("\nCorrect!\n")
            numCorrectGuesses += 1
            # find all instances of letter in word
            charCounter = 0
            for char in guessWord:
              if char == guessLetter:
                correctLetters[charCounter] = guessLetter
              charCounter += 1
          else:
            print("\nWrong :/\n")
            numIncorrectGuesses += 1
            incorrectLetters.append(guessLetter)
          # output guessed letters and number of guesses
          print(" ".join(correctLetters) + "\n")
          print(f"Correct Guesses:\t{numCorrectGuesses}")
          print(f"Incorrect Guesses:\t{numIncorrectGuesses}\t({', '.join(incorrectLetters)})\n")
        else:
          print("\nYou already guessed that letter!\nTry again.\n")
      else:
        print("\nYou can only guess letters in the alphabet\n")
    else:
      print("\nYou can only guess one letter.\nTry again.\n")
    
    # check if word is complete
    if guessWord == "".join(correctLetters):
      print(f"Great job! You guessed the correct word with {numCorrectGuesses + numIncorrectGuesses} guesses.")
      guessContinue = False

  # play again?
  continueInput = input("\nWould you like to play again? (If so type 'Y'):\t")
  if continueInput.upper() != "Y":
    gameContinue = False

# say goodbye to user
print("\nThanks for playing!")
```
